```python
import requests
import pandas as pd
import datetime
import ast
```

```python
access_token = ''
group_name = 'hynta_atmta' #id или короткое имя (которое в адресной строке) паблика
score = 0 #0-100 0-совсем не подходит 100 идеально подходит
```

```python
#Документация по groups.getById     https://dev.vk.com/method/groups.getById
#Документация по полям groups       https://dev.vk.com/reference/objects/group
group = requests.get('https://api.vk.com/method/groups.getById?'+
                       'group_id='+group_name+'&'+
                       'v=5.131&'+
                        'fields=activity,age_limits,ban_info,members_count,trending,verified,wall&'+
                       'access_token='+access_token).json()['response'][0]
print(group)
```

```python
is_closed = group['is_closed'] #0 — открытое; 1 — закрытое; 2 — частное. | очевидно нужно 0 
activity = group['activity'] #Строка тематики паблика. |куча разных, выбирать под тип рекламы (реклама магазина техники - тематика "техника")
age_limits = group['age_limits'] #1 — нет; 2 — 16+; 3 — 18+ | предпочтительно 1
members_count = group['members_count'] #число подписчиков | чем больше - чем лучше
trending = group['trending'] #Информация о том, есть ли у сообщества «огонёк». | больше охват если есть 
verified = group['verified'] #1 — является; 0 — не является. | больше охват 
wall = group['wall']#0 — выключена;1 — открытая;2 — ограниченная;3 — закрытая. | если 1 то много мусора но можно и самому бесплатно оставлять рекламу
print(is_closed,activity,age_limits,members_count,trending,verified,wall)
```

```python
#Документация по groups.getMembers  https://dev.vk.com/method/groups.getMembers
next_from = '0'
mass = []
count = 0
while next_from != '':
  print('Обработка пользователей от ' + str(count*1000)+ ' до ' + str((count+1)*1000-1))
  group_members = requests.get('https://api.vk.com/method/groups.getMembers?'+
                       'group_id='+group_name+'&'+
                       'v=5.131&'+
                       'fields=deactivated&'
                       'start_from=' + next_from + 
                       '&access_token='+access_token).json()['response']
  mass.extend(group_members['items'])
  next_from = group_members['next_from']
  count = count + 1
```

```python
#документация по users.get     https://dev.vk.com/method/users.get
#документация по полям users   https://dev.vk.com/reference/objects/user
next_from = '0'
user_info = []
count = 0
while count <= len(mass)//1000 + 1:
  print('Обработка пользователей от ' + str(count*1000)+ ' до ' + str((count+1)*1000-1))
  group_members = requests.get('https://api.vk.com/method/users.get?'+
                       'user_ids='+str(mass[count*1000:(count+1)*1000-1])+'&'+
                       'v=5.131&'+
                       'fields=bdate,city,followers_count,last_seen,relation,personal,sex,trending,verified&'
                       '&access_token='+access_token).json()['response']
  user_info.extend(group_members)
  count = count + 1
```

```python
df_user_info = pd.DataFrame.from_records(user_info)
```

```python
df_user_info_deleted_closed_profiles = df_user_info.drop(df_user_info[df_user_info.is_closed == True].index)
df_user_info_deleted_closed_profiles = df_user_info.drop(['first_name','last_name','can_access_closed','is_closed','relation_partner'],axis=1)
```

```python
df_user_info_deleted_closed_profiles['bdate'] = df_user_info_deleted_closed_profiles.apply(lambda x:
                                                                                           int(str(x['bdate']).split('.')[2]) if len(str(x['bdate'])) > 5 else None,
                                                                                           axis=1)
df_user_info_deleted_closed_profiles['city'] = df_user_info_deleted_closed_profiles.apply(lambda x:
                                                                                           str(str(x['city']).split('title')[1].split('\'')[2]) if type(x['city']) != float else None,
                                                                                           axis=1)
df_user_info_deleted_closed_profiles['last_seen'] = df_user_info_deleted_closed_profiles.apply(lambda x: 
                                                                                            datetime.datetime.fromtimestamp(int(str("{'platform': 4, 'time': 1685029760}").split('time')[1][3:-1])),
                                                                                              axis=1)
df_user_info_deleted_closed_profiles['alcohol'] = df_user_info_deleted_closed_profiles.apply(lambda x: 
                                                                                            (ast.literal_eval(str(x['personal']))).get('alcohol') if len(str(x['personal'])) > 5 else None,
                                                                                              axis=1)
df_user_info_deleted_closed_profiles['life_main'] = df_user_info_deleted_closed_profiles.apply(lambda x: 
                                                                                            (ast.literal_eval(str(x['personal']))).get('life_main') if len(str(x['personal'])) > 5 else None,
                                                                                              axis=1)
df_user_info_deleted_closed_profiles['people_main'] = df_user_info_deleted_closed_profiles.apply(lambda x: 
                                                                                            (ast.literal_eval(str(x['personal']))).get('people_main') if len(str(x['personal'])) > 5 else None,
                                                                                              axis=1)
df_user_info_deleted_closed_profiles['political'] = df_user_info_deleted_closed_profiles.apply(lambda x: 
                                                                                            (ast.literal_eval(str(x['personal']))).get('political') if len(str(x['personal'])) > 5 else None,
                                                                                              axis=1)
df_user_info_deleted_closed_profiles['smoking'] = df_user_info_deleted_closed_profiles.apply(lambda x: 
                                                                                            (ast.literal_eval(str(x['personal']))).get('smoking') if len(str(x['personal'])) > 5 else None,
                                                                                              axis=1)
df_user_info_deleted_closed_profiles['religion'] = df_user_info_deleted_closed_profiles.apply(lambda x: 
                                                                                            (ast.literal_eval(str(x['personal']))).get('religion') if len(str(x['personal'])) > 5 else None,
                                                                                              axis=1)
df_user_info_deleted_closed_profiles = df_user_info_deleted_closed_profiles.drop(['personal'],axis=1)
```
